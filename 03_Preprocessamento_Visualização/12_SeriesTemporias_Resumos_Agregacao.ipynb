```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('SeriesTemporais_Python/Download/Dados/avocado.csv', parse_dates=['Date'])
```

```python
data.head()
```

```python
# agrupado o valor total por data
soma = data.groupby('Date')['Total Volume'].mean().reset_index()
soma.head()
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,6))
soma.set_index('Date').plot(ax=ax,marker='o', linestyle='--',color='blue')
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,6))
(soma.set_index('Date')
 .assign(month=lambda df: df.index.month)
 .groupby('month')['Total Volume'].agg(['mean','std','median','min','max'])
 .plot(ax=ax,marker='o', linewidth=1))
ax.set_xlabel('Mês')
ax.set_ylabel('Volume')
```
