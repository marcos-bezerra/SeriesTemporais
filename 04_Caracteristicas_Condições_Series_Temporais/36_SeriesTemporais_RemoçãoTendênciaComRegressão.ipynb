```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15,6
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from statsmodels.tsa.stattools import adfuller
from pandas import Series
```

```python
data = pd.read_csv('/Users/marcos/Documents/00_Git_Projetos/SeriesTemporais/SeriesTemporais_Python/Download/Dados/AirPassengers.csv',
                   parse_dates=['Month'], index_col='Month') #, date_parser=dateparse)
```

```python
X = [i for i in range(0, len(data))]
X = np.reshape(X, (len(X),1))
y = data.values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
trend = model.predict(X)
```

```python
plt.plot(y)
plt.plot(trend)
plt.show()
```

série temporal com a linha de tendência

```python
detrended = [y[i]-trend[i] for i in range(0,len(data))]
```

```python
plt.plot(detrended)
```
