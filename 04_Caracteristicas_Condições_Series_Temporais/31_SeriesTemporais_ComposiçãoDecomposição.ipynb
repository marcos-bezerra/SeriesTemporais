```python
plt.plot(data)
plt.title('Passageiros Internacionais Transportados', fontsize=24)
plt.ylabel('Passageiros')
plt.xlabel('Data')
plt.show()
```

```python
# Pacote para poder decompor
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# decomposio modelo MULTIPLICATIVO
result = seasonal_decompose(data, model='multiplicative')
result.plot()
plt.show()
```

```python
# decomposio modelo ADITIVE
result2 = seasonal_decompose(data, model='aditive')
result2.plot()
plt.show()
```
